While pandas has many built In methods,
we can use the .apply() method call to
apply any custom python function of our
own to every row in a Series.

We can use either one or multiple columns
as input, let's explore this in the notebook!

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

```python
#To get the last four digits of the CC Number column
str(3560325168603410)[-4:]
```

```python
# We can define a function that works the same as the above code
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(3560325168603410)
```

```python
df['last_four']=df['CC Number'].apply(last_four)
```

The above line applies last_four() function to all the values of CC Number column

```python
df
```

```python
df['total_bill'].mean()
```

```python
def any(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['any'] = df['total_bill'].apply(any)
```

```python
df['any']
```

## The functions in the apply method must return only a single value not a series

```python
# def simple(num):
#     return num*2
# The above function can be written using lamda
# lambda num: num*2
```

## The lamba cannot be used for complex expressions

```python
df['total_bill'].apply(lambda bill: bill*2)
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous "
    else:
        return "Other"
```

```python
quality(16.99,1.01)
```

```python
 df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1) 
```

1. Select the columns on which you want to apply 
2. define the dataframe name eg. df:
3. define function and pass the parameters 

```python
 df['quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1) 
```

```python
df
```

Another method to apply a complex function to multiple columns

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df.head()
```

## Check the time taken by both the methods using the timeit module

```python
import timeit
```

```python
 # code snippet to be executed only once
setup =  '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
 def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
 '''
```

```python
#Code snippet whose execution time is to be measured

stmt_one ='''
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1) 
'''
stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
'''
```

```python
timeit.timeit(setup=setup,stmt=stmt_one,number=1000)
```

```python
timeit.timeit(setup=setup,stmt=stmt_two,number=1000)
```
