```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

# Adding Legend to the plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X^2')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X^2')
ax.legend(loc="lower right")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X^2')
ax.legend(loc=(1,0.5))
```

# Adding color to plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='red')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0') # can also use HEX color code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0') 
ax.plot(x,x+1,color='#29d7e3')
```

# Line width

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0',lw=4) 
ax.plot(x,x+1,color='#29d7e3',linewidth='8')
```

# Line style

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0',lw=4,ls='--') 
ax.plot(x,x+1,color='#29d7e3',linewidth='8',ls='-.')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,x+1,color='#29d7e3',linewidth='8')
lines[0].set_dashes([5,2,10,2]) 
# (5 size wide line,2 spaces,10 width line,2 spaces)
```

# Marker

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0',lw=2,marker='o',ls='--') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0',lw=2,marker='o',ls='--',markersize=10) 
ax.plot(x,x+1,color='red',lw=1,marker='+',ls='--',ms=14) 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#ab56b0',lw=2,marker='o',ls='--',markersize=15,markerfacecolor='red',markeredgewidth=8,markeredgecolor='orange') 
```
