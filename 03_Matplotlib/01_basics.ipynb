```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = 2 * x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X axis") #Put a label at x-axis
plt.ylabel("Y axis") #Put a label at y-axis
plt.xlim(0,6) # set a limit at x-axis
plt.ylim(0,15)# set a limit at y-axis
plt.title("String title"); # To add a title 
plt.savefig("myfirstplot.png") # saves the plot in the current directory
#If you want to save the plot in a different location then just provide the path
```
