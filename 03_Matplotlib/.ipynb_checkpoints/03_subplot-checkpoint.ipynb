```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[1].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)
axes[0].plot(x,y)
axes[1].plot(a,b)
axes[2].plot(x,y)
plt.tight_layout() #To make space between the multiple plots
```

```python
#If there are many plots of different axes then to make proper space manually between them use subplots_adjust() method
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(x,y)
fig.subplots_adjust(wspace=0.5,hspace=0.5)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel("Aalam")
axes[0][0].set_ylabel("Asad")
axes[0][0].set_title("Friends meter")

axes[0][1].plot(a,b)


axes[1][0].plot(a,b)
axes[1][0].set_xlabel("Awaiz")
axes[1][0].set_ylabel("Fahad")
axes[1][0].set_title("Bestbros meter")

fig.suptitle("Figure level title")

axes[1][1].plot(x,y)

fig.tight_layout()

fig.savefig("Third_plot",bbox_inches="tight")
```
