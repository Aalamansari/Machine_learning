```python
#Create the figure object 
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) # providing the axes in the plot
axes.plot(a,b)
plt.show()
```

```python
#Create the figure object 
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1]) # providing the axes in the plot
axes1.plot(a,b)

#SMALLER AXES
axes2 = fig.add_axes([0.5,0.5,0.2,0.2]) # providing the axes in the plot
axes2.plot(x,y)
plt.show()
```

```python
#Create the figure object 
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1]) # providing the axes in the plot
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power")

#SMALLER AXES
axes2 = fig.add_axes([0.5,0.5,0.2,0.2]) # providing the axes in the plot
axes2.plot(x,y)
plt.show()
```

```python
#Create the figure object 
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1]) # providing the axes in the plot
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of A")

#SMALLER AXES
axes2 = fig.add_axes([0.5,0.5,0.2,0.2]) # providing the axes in the plot
axes2.plot(x,y)
axes2.set_xlim(1,5)
axes2.set_ylim(1,8)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Power of X")
plt.show()
```

```python
 #To make the plot look more clear use the dpi attribute
fig = plt.figure(dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
 #To change the plot size use the figsize attribute
fig = plt.figure(figsize=(2,2),dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig = plt.figure(figsize=(2,2),dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig("second_plot.png",bbox_inches="tight") # bbox_inches = "tight" is for making the axes visible in the saved plot
```
