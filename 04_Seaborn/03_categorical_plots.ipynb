```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df
```

```python
df['division'].value_counts()
```

# Countplot

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division',palette='Set2')
```

# Barplot

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df
```

# Boxplot

```python
plt.figure(figsize=(10,3),dpi=200)
sns.boxplot(data=df,y='math score',x='test preparation course')
```

```python
plt.figure(figsize=(10,3),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education')
```

```python
# hue helps to create sub-levels in the plot
plt.figure(figsize=(10,3),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
# hue helps to create sub-levels in the plot
plt.figure(figsize=(10,7),dpi=200)
sns.boxplot(data=df,x='math score',y='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,7),dpi=200)
sns.boxplot(data=df,x='math score',y='parental level of education',hue='test preparation course',palette='Set2')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

# Violin plot

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',hue='test preparation course')
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',hue='test preparation course',split=True)
plt.legend(bbox_to_anchor=(1.25,0.5))
```

```python
#If you want to erase the middle line from the plot then use inner=None
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',hue='test preparation course',split=True,inner=None)
plt.legend(bbox_to_anchor=(1.25,0.5))
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',split=True,inner='quartile')
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',split=True,inner='sticks')
```

```python
# Adjust the bandwidth using the bw 
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',split=True,inner='quartile',bw=0.1)
```

# Swarmplot

```python
sns.swarmplot(data=df,x='math score')
# The warning comes because the figure is not able to take all the points
```

```python
sns.swarmplot(data=df,x='math score',size=2)
```

```python
sns.swarmplot(data=df,x='math score',y='gender',size=2)
```

```python
sns.swarmplot(data=df,x='math score',y='gender',hue='test preparation course',size=2,dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
# dodge makes the more clear
```

# Boxenplot

```python
sns.boxenplot(x='math score',y='test preparation course',data=df,hue='gender')
```
