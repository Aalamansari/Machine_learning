```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df
```

```python
df = df.set_index('Countries')
```

```python
df
```

# Heatmap

```python
sns.heatmap(df)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1))
#Dropping the life expectancy cause it is the only column with different type of values in comparsion to other cols
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```

```python
plt.figure(dpi=300)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```

```python
# Changing the color mapping of the histogram with cmap
plt.figure(dpi=300)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,cmap='viridis')
```

# Clustermap

```python
plt.figure(dpi=300)
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,cmap='viridis')
```

```python
plt.figure(dpi=300)
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,cmap='viridis',
              col_cluster=False)
```
