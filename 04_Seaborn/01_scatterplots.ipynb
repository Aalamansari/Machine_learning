```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df
```

```python
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
#hue means to color the points of the defined column
# level of education is categorical column
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')
```

```python
# salary is continuous column
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')
```

```python
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2')
```

```python
# Points of a column can be displayed in different sizes using size 
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,size='level of education')
```

```python
#To give the same size to all the points in the plot use s
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,s=150)
```

```python
# To define the transparency of the plot use alpha
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,s=150,alpha=0.2)
```

```python
#To make the different values of the categorical column represented by different shapes use style
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,s=200,style='level of education')
```

```python
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,s=200,style='level of education',hue='level of education')
```

```python
plt.figure(figsize=(12,5),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,s=200,style='level of education')
plt.savefig("first_plot.jpeg")
```
