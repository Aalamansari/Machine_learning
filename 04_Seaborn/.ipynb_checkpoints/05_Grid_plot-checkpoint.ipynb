```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch')
plt.show()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',hue='lunch')
plt.show()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch')
plt.show()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch',col='test preparation course')
plt.show()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='violin',row='lunch',col='test preparation course')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

## Edit the pairplot as per our requirement

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)

plt.show()
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()

plt.show()
```
