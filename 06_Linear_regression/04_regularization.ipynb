```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
X = df.drop('sales',axis=1)
```

```python
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features = polynomial_converter.fit_transform(X)
```

```python
X.shape
```

```python
poly_features.shape # 19 column are the new features generated by polynomial regression of degree 3
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
# Now scale the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# fitting the scaler to only training set to avoid the data leakage scenario
scaler.fit(X_train)
```

```python
# new scaled data
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

# Ridge regression

```python
from sklearn.linear_model import Ridge
```

```python
# help(Ridge)
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
RMSE
```

```python
from sklearn.linear_model import RidgeCV
# RidgeCV performs a variety of cross validation for a multiple alpha values and report backs the best alpha value
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1,1.0,10),scoring='neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
# To check the best possible value of alpha
```

```python
from sklearn.metrics import SCORERS
```

```python
SCORERS.keys()
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
RMSE
```

```python
ridge_cv_model.coef_
```

## LASSO - least absolute shrinkage and selection operator

```python
# first step is to check for a suitable alpha value
```

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(eps=0.001,n_alphas=100,cv=5,max_iter=100000)
```

```python
lasso_cv_model.fit(X_train,y_train)
```

```python
 lasso_cv_model.alpha_
```

```python
 test_predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
RMSE
```
